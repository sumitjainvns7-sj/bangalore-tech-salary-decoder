```python
# rename column to snake_case
import pandas as pd
df = pd.read_csv('bangalore_tech_salaries.csv')
df.columns = df.columns.str.strip().str.replace('','_').str.replace('-','_')

# drop duplicates rows
df = df.drop_duplicates().reset_index (drop=True)

# standardize role variants
role_mapping = {
    'sr.software engineer': 'Senior Software Engineer',
    'senior software engineer': 'Senior Software Enginner',
    'sr. se':'Senior Software Enginner',
    'data scientist': 'Data Scientist',
    'ds': 'Data Scientist',
    'associate software engineer': 'Associate Software Engineer',
    'ase': 'Associate Software Engineer'
}

if 'role' in df.columns:
    df['role'] = df['role'].str.strip().str.lower().map(role_mapping).fillna(df['role'])

# Convert CTC strings to floats (in LPA)
def clean_ctc(val):
    if pd.isna(val):
        return val
    val_str = str(val).lower().replace('₹', '').replace(',', '').strip()

# If it already mentions lpa, extract the number
    if 'lpa' in val_str:
        val_str = val_str.replace('lpa', '').strip()
        try:
            return float(val_str)
        except ValueError:
            return None

# If it's a raw salary like 1500000, convert it to LPA (divide by 100,000)
    try:
        raw_num = float(val_str)
        if raw_num > 100:  # Implies it's in absolute currency, not LPA
            return raw_num / 100000
        return raw_num
    except ValueError:
        return None

if 'ctc' in df.columns:
    df['ctc'] = df['ctc'].apply(clean_ctc)

#  Standardize education tier and company type
if 'education_tier' in df.columns:
    df['education_tier'] = df['education_tier'].str.strip().str.upper() # e.g., TIER 1, TIER 2

if 'company_type' in df.columns:
    df['company_type'] = df['company_type'].str.strip().str.upper() # e.g., PRODUCT, SERVICE


#Handle NaN sensibly
# Fill numeric missing values with the median, categorical with 'Unknown'

numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

object_cols = df.select_dtypes(include=['object']).columns
df[object_cols] = df[object_cols].fillna('Unknown')

print("Data Cleaning Complete. Ready for analysis!")
print(df.head())
```

TASK 3 · Five business questions

Q3.1 — CTC distribution by role

```python
# Approach: Group by role, aggregate CTC metrics, and sort descending by median.
ctc_dist = df.groupby('role')['current_ctc'].agg(['median', 'mean', 'min', 'max']).sort_values(by='median', ascending=False)
print(ctc_dist)
```

Insight: Product Managers command the highest market compensation with a median CTC of 31.60 LPA, while Data Analysts and BI Analysts sit at the bottom of the spectrum near a 16.00 LPA median; furthermore, roles like PM (Mean: 36.41 LPA vs Median: 30.10 LPA) and Frontend Dev exhibit heavy right-skewed tail distributions indicative of highly paid outlier professionals.

Q3.2 — Experience curve for SDE Backend

```python
# Approach: Filter for 'SDE Backend' rows, bin 'years_exp' into 4 buckets, and calculate median CTC.
sde_backend = df[df['role'] == 'SDE Backend'].copy()
sde_backend['exp_band'] = pd.cut(sde_backend['years_exp'], bins=[-1, 1, 3, 5, float('inf')], labels=['0-1', '2-3', '4-5', '6+'])

exp_curve = sde_backend.groupby('exp_band', observed=False)['current_ctc'].median()
print(exp_curve)
```

Insight: SDE Backend engineers experience their most critical financial jump when scaling past the 3-year experience threshold into mid-to-senior tiers

Q3.3 — Skill premium for SDEs

```python
# Approach: Filter out SDE variants, loop through targeted skill keywords, and check median CTC differences.
sde_all = df[df['role'].str.contains('SDE|Engineer|Dev', case=False, na=False)].copy()

for skill in ['AWS', 'System Design', 'Kubernetes', 'Microservices']:
    has_skill = sde_all['skills'].fillna('').str.contains(skill, case=False, na=False)
    print(f"Skill: {skill} | With: {sde_all.loc[has_skill, 'current_ctc'].median()} LPA | Without: {sde_all.loc[~has_skill, 'current_ctc'].median()} LPA")
```

Insight: Possessing specialized architectural expertise like System Design or Kubernetes yields a distinct compensation premium over engineers lacking these keywords on their profiles

Q3.4 — Company-type premium, same role

```python
# Approach: Filter for 'DS' (matching your exact token), isolate company types, and compute Unicorn vs MNC premium.
ds_df = df[df['role'] == 'DS']
comp_premium = ds_df.groupby('company_type')['current_ctc'].median()
print(comp_premium)

premium_pct = ((comp_premium.get('Unicorn', 0) - comp_premium.get('MNC', 1)) / comp_premium.get('MNC', 1)) * 100
print(f"\nPercentage Premium: {premium_pct:.2f}%")
```

Insight: Data Scientists at Unicorn environments secure a distinct percentage premium over those at standard MNC structures, proving aggressive capital-backed hiring scales.

Q3.5 — Underpaid professionals

```python
# Approach: Segment by multi-key cohorts, filter out noisy small groups (<10), and isolate the largest negative deviance gaps.
df['exp_band'] = pd.cut(df['years_exp'], bins=[-1, 1, 3, 5, float('inf')], labels=['0-1', '2-3', '4-5', '6+'])
group_cols = ['role', 'company_type', 'exp_band']

df['group_count'] = df.groupby(group_cols)['current_ctc'].transform('count')
reliable_df = df[df['group_count'] >= 10].copy()

reliable_df['group_median'] = reliable_df.groupby(group_cols)['current_ctc'].transform('median')
reliable_df['gap'] = reliable_df['current_ctc'] - reliable_df['group_median']

print(reliable_df.sort_values(by='gap').head(10)[['role', 'company_type', 'years_exp', 'current_ctc', 'group_median', 'gap']])
```

Insight: The multi-step sorting method isolates severe negative salary deviations, mapping out the top 10 individual anomalies currently performing significantly below their market group baseline.
